# Importando librerías

```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

# Ejercicio 1: Graficar variable de Temperatura

```python
Temperatura = ctrl.Antecedent(np.arange(80, 110, 1), 'Temperatura')
Temperatura['High'] = fuzz.trapmf(Temperatura.universe, [80, 80, 90, 100])
Temperatura['High High'] = fuzz.trapmf(Temperatura.universe, [90, 100, 110, 110])

Temperatura.view()
```
